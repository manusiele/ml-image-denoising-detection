# Pascal VOC Dataset Exploration
Explore the Pascal VOC dataset structure and visualize samples

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import xml.etree.ElementTree as ET

%matplotlib inline
```

```python
# Dataset paths
DATA_DIR = Path('../data/pascal-voc')
print(f"Dataset directory: {DATA_DIR}")
print(f"Exists: {DATA_DIR.exists()}")
```

```python
# List dataset structure
for item in DATA_DIR.rglob('*'):
    if item.is_dir():
        print(f"📁 {item.relative_to(DATA_DIR)}")
```

```python
# VOC classes
VOC_CLASSES = [
    'aeroplane', 'bicycle', 'bird', 'boat', 'bottle',
    'bus', 'car', 'cat', 'chair', 'cow',
    'diningtable', 'dog', 'horse', 'motorbike', 'person',
    'pottedplant', 'sheep', 'sofa', 'train', 'tvmonitor'
]
print(f"Number of classes: {len(VOC_CLASSES)}")
print(f"Classes: {VOC_CLASSES}")
```
