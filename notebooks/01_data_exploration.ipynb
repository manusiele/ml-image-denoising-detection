# Pascal VOC Dataset Exploration
Explore the Pascal VOC dataset structure and visualize samples

```python
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET

%matplotlib inline
```

```python
# Import Kaggle data loader
sys.path.append('../src')
from data.kaggle_loader import KaggleDataLoader

# Initialize Kaggle API loader
loader = KaggleDataLoader('huanghanchina/pascal-voc-2012')
print("✓ Connected to Kaggle API")
```

```python
# List available files in dataset
files = loader.list_files()
print(f"Total files in dataset: {len(files)}")
print("\nSample files:")
for f in files[:10]:
    print(f"  📄 {f}")
```

```python
# VOC classes
VOC_CLASSES = [
    'aeroplane', 'bicycle', 'bird', 'boat', 'bottle',
    'bus', 'car', 'cat', 'chair', 'cow',
    'diningtable', 'dog', 'horse', 'motorbike', 'person',
    'pottedplant', 'sheep', 'sofa', 'train', 'tvmonitor'
]
print(f"Number of classes: {len(VOC_CLASSES)}")
print(f"Classes: {VOC_CLASSES}")
```

## Visualize Sample Images

```python
# Load and visualize sample images from Kaggle
print("Loading sample images from Kaggle API...")

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, (filename, img) in enumerate(loader.stream_dataset_files('*.jpg', max_files=6)):
    axes[idx].imshow(img)
    axes[idx].set_title(filename.split('/')[-1])
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```
